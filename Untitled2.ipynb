# **Clean the Data**

### **Step 1: Drop Duplicates**

```python
df.drop_duplicates(inplace=True)
print("Shape of DataFrame after dropping duplicates:", df.shape)
```

### **Step 2: Drop Columns with Insufficient Data**

```python
df.drop(['house_rules', 'license'], axis=1, inplace=True)
print("Shape of DataFrame after dropping columns:", df.shape)
```

### **Step 3: Clean Price and Service Fee Columns**

```python
# Remove '$' and ',' from 'price' and 'service fee' columns and convert to numeric
df['price'] = df['price'].astype(str).str.replace(r'[$,]', '', regex=True)
df['service fee'] = df['service fee'].astype(str).str.replace(r'[$,]', '', regex=True)

df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['service fee'] = pd.to_numeric(df['service fee'], errors='coerce')

print("Data types after cleaning price and service fee:")
display(df[['price', 'service fee']].dtypes)
```

### **Step 4: Rename Price and Service Fee Columns**

```python
df.rename(columns={'price$': 'price_$', 'service fee$': 'service_fee_$'}, inplace=True)
print("Columns after renaming:")
display(df.columns)
```

### **Step 5: Drop Rows with Missing Values**

```python
df.dropna(inplace=True)
print("Shape of DataFrame after dropping rows with missing values:", df.shape)
```

### **Step 7: Correct Spelling**

```python
df['neighbourhood group'] = df['neighbourhood group'].replace('brookln', 'Brooklyn')
print("Value counts for 'neighbourhood group' after correcting spelling:")
display(df['neighbourhood group'].value_counts())
```

### **Step 8: Handle Outliers**

```python
# For demonstration, let's use the IQR method to identify and remove outliers in 'availability 365'
Q1 = df['availability 365'].quantile(0.25)
Q3 = df['availability 365'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['availability 365'] >= lower_bound) & (df['availability 365'] <= upper_bound)]

print("Shape of DataFrame after handling outliers in 'availability 365':", df.shape)
```

### **Step 9: Finish task**

```python
print("Cleaned DataFrame head:")
display(df.head())

print("\nCleaned DataFrame Info:")
display(df.info())
```

## Conclusion of Exploratory Data Analysis (EDA)

Based on the exploratory data analysis performed on the Airbnb dataset, here are some key findings:

**Data Cleaning:**
*   We successfully identified and removed duplicate records, resulting in a cleaner dataset for analysis.
*   Columns with insufficient data, such as 'house_rules' and 'license', were dropped.
*   The 'price' and 'service fee' columns were cleaned by removing currency symbols and commas and converting them to a numeric data type.
*   Missing values were handled by dropping rows with null entries after the initial cleaning steps.
*   A spelling correction was applied to the 'neighbourhood group' column.
*   Outliers in the 'availability 365' column were addressed using the IQR method.

**Property Types:**
*   The dataset contains four main property types: 'Entire home/apt', 'Private room', 'Shared room', and 'Hotel room'.
*   'Entire home/apt' and 'Private room' are the most common property types, significantly outnumbering 'Shared room' and 'Hotel room'.

**Neighbourhood Analysis:**
*   We identified the neighbourhood with the highest number of listings and visualized the top 10 neighbourhoods by listing count. This showed significant variation in listing density across different neighbourhoods.
*   Analysis of average prices by neighbourhood group revealed differences in pricing across different areas, with some groups having slightly higher average prices than others.

**Price and Service Fee Relationship:**
*   A regression plot showed a positive correlation between the price of a listing and its service fee, suggesting that higher-priced listings tend to have higher service fees.

**Host Analysis:**
*   We identified the top hosts based on their calculated host listings count.
*   An analysis of average review rates by host identity verification showed a very slight difference in average review rates between verified and unconfirmed hosts, but a box plot provided a clearer picture of the distribution, indicating similar overall distributions with some differences in outliers or spread.
*   A regression plot between 'calculated host listings count' and 'availability 365' showed a weak positive correlation, suggesting that hosts with more listings may have slightly higher availability on average, but the relationship is not strong.

**Construction Year and Price:**
*   A line plot of the average price by construction year showed fluctuations in average price across different construction years, but no clear strong linear trend was immediately apparent.

Overall, the EDA has provided valuable insights into the distribution of property types and listings across neighbourhoods, pricing trends, host behavior, and relationships between key variables in the Airbnb dataset. These findings can inform further analysis and potential modeling.
